# game.py(전체 갱신)

* 게임 상태

게임 상태 'State'를 정의한다.

1. __init__(pieces=None, enemy_pieces=None)
: 게임 상태 초기화

'State' 생성자는 게임 상태를 초기화한다. 자신의 돌의 배치, 상대방의 돌의 위치를 길이 42(7열 * 6행)의 1차원 배열 'pieces', 'enemy_pieces'에 저장한다. 커넥트4에서 돌의 초기 배치는 모두 빈 공간이다.

2. piece_count(pieces)
: 돌의 수 취득

무승부를 판단할 때 사용한다.

3. is_lose()
: 패배 여부 판정

4. is_draw()
: 무승부 여부 판정

5. is_done()
: 게임 종료 여부 판정

6. next(action)
: 다음 상태 취득

'next(aciton)'은 행동에 따른 다음 상태를 얻는다. 돌의 배치 'pieces'를 복사한 뒤, 지정한 열에서 비어 있는 위치 중 가장 아래(T 좌표가 가장 큰) 장소에 돌을 배치한다.


7. legal_actions()
: 둘 수 있는 수 취득

'legal_action()'은 둘 수 있는 수 리스트를 얻는다. 커넥트4에서의 둘 수 있는 수는 가장 위의 행(pieces와 enemy_pieces의 0~6)에서 돌이 가득 차 있지 않은 매스다.

8. is_first_player()
: 선 수 여부 취득

9. __str__()
: 문자열 표시
